```python
class ThreeNode:

    def __init__(self, data = None):
        self._data = data 
        self._children = []

    def add_child(self, child_node):
        self._children.append(child_node)
    
    def remove_child(self, child_node):
        for i in len(self._children):
            if self._children[i] is child_node:
                del(self._children[i])
                break
    
    def get_children(self):
        return self._children

    def get_data(self):
        return self._data
    
    def set_data(self, data):
        self._data = data
    
```

```python
###binary search tree::

class BSTNode:

    def __init__(self, data = None , left = None , right = None):
        self._data = data
        self._left = left
        self._right = right

    def add_node(self, data):
        current_node = self
        while True :
            if data <= current_node._data:
                if current_node._left is None:
                    current_node._left = BSTNode(data)
                    return current_node._left
                else:
                    current_node = current_node._left
            else:
                if current_node._right is None :
                    current_node._right = BSTNode(data)
                    return current_node._right
                else:
                    current_node = current_node._right
            

    def search_node(self, data_to_search):
        if self._data == data_to_search :
            return self
        if data_to_search < self._data:
            if self._left is None:
                return None 
            return self._left.search_node(data_to_search)
        else:
            if self._right is None :
                return None
            return self._right.search_node(data_to_search)
        
    
```

```python
#implementazione della clase grafo :: ---- fatta con me e copylot ----

class Graph:

    def __init__(self):
        self._nodes = {}


    def add_node(self, n):
        if n not in self._nodes:
            self._nodes[n] = { }

    def remove_node(self, n):
        if n not in self._nodes:
            self._nodes.pop(n)

    def get_nodes(self):
        return self._nodes.keys()
    
    def add_edge(self, n1, n2, w = 1):
        if n1 not in self._nodes:
            self._nodes[n1] = { }
        if n2 not in self._nodes:
            self._nodes[n2] = { }
        self._nodes[n1][n2] = w

    def remove_edge(self, n1, n2):  
        if n1 in self._nodes and n2 in self._nodes[n1]:
            self._nodes[n1].pop(n2) 
    
    def get_edges(self):    
        edges = []
        for n1 in self._nodes:
            for n2 in self._nodes[n1]:
                edges.append((n1, n2, self._nodes[n1][n2]))
        return edges        
    
    def get_neighbors(self, n):   
        return self._nodes[n].keys()    
    
    def get_weight(self, n1, n2):   
        return self._nodes[n1][n2] if n2 in self._nodes[n1] else None   
    
    def is_adjacent(self, n1, n2):   
        return n2 in self._nodes[n1] if n1 in self._nodes else False    
    
    def __str__(self):   
        return str(self._nodes) 
    
    def __repr__(self): 
        return str(self._nodes)

    
#implementazione della classe grafo con matrice di adiacenza::
    
class GraphMatrix:

    def __init__(self):
        self._nodes = []
        self._edges = []
        self._matrix = []   

    def add_node(self, n):
        if n not in self._nodes:
            self._nodes.append(n)
            for row in self._matrix:
                row.append(0)
            self._matrix.append([0] * len(self._nodes))

    def remove_node(self, n):
        if n in self._nodes:
            index = self._nodes.index(n)
            self._nodes.pop(index)
            self._matrix.pop(index)
            for row in self._matrix:
                row.pop(index)


    def get_nodes(self):
        return self._nodes
    
    def add_edge(self, n1, n2, w = 1):
        if n1 in self._nodes and n2 in self._nodes:
            index1 = self._nodes.index(n1)
            index2 = self._nodes.index(n2)
            self._matrix[index1][index2] = w
            self._matrix[index2][index1] = w
            self._edges.append((n1, n2, w))

    def remove_edge(self, n1, n2):
        if n1 in self._nodes and n2 in self._nodes:
            index1 = self._nodes.index(n1)
            index2 = self._nodes.index(n2)
            self._matrix[index1][index2] = 0
            self._matrix[index2][index1] = 0
            self._edges.remove((n1, n2, self._matrix[index1][index2]))

    def get_edges(self):
        return self._edges
    
    def get_neighbors(self, n):
        if n in self._nodes:
            index = self._nodes.index(n)
            neighbors = []
            for i in range(len(self._matrix[index])):
                if self._matrix[index][i] != 0:
                    neighbors.append(self._nodes[i])
            return neighbors
    
    def get_weight(self, n1, n2):
        if n1 in self._nodes and n2 in self._nodes:
            index1 = self._nodes.index(n1)
            index2 = self._nodes.index(n2)
            return self._matrix[index1][index2]
        return None
    
    def is_adjacent(self, n1, n2):
        if n1 in self._nodes and n2 in self._nodes:
            index1 = self._nodes.index(n1)
            index2 = self._nodes.index(n2)
            return self._matrix[index1][index2] != 0
        return False
    
    def __str__(self):
        return str(self._matrix)
    
    def __repr__(self):
        return str(self._matrix)
    
    


```

```python
####implementazione algoritmo di dijkstra ::

import numpy as np


def dijkstra(ad_matrix, start_node):
    n = len(ad_matrix)
    unvisited_nodes = set(range(n))
    previous_nodes ={}
    path_weights = np.full(n, float('inf'))
    path_weights[start_node] = 0

    while unvisited_nodes:
        
        current_min_node = None

        for node in unvisited_nodes:
            if current_min_node is None: 
                current_min_node = node
            elif path_weights[node] < path_weights[current_min_node]: 
                current_min_node = node

        unvisited_nodes.remove(current_min_node)

        for node in unvisited_nodes:
            if ad_matrix[node,current_min_node]:
                # the node is a neighbor
                weight = path_weights[current_min_node]+ad_matrix[node,current_min_node]    
                if weight < path_weights[node]:
                    path_weights[node]=weight 
                    previous_nodes[node]=current_min_node
    return path_weights, previous_nodes

```
